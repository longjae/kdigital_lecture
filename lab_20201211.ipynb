```python
#성적관리 프로그램 개발
menu = -1
score_list = []

while int(menu) != 0:
    menu = input("메뉴를 선택해주세요 (1 - 입력, 2 - 조회, 3 - 삭제, 0 - 종료) : ")

    if int(menu) == 1:
        name = input("이름 : ")
        math = input("수학 : ")
        science = input("과학 : ")
        eng = input("영어 : ")     

        temp = []
        
        keys = ["이름","수학","과학","영어"]
        users = [name,math,science,eng]
        dic = dict(zip(keys,users))
        temp.append(dic)    
        score_list.append(temp)
            
    elif int(menu) == 2:
        for idx, val in enumerate(score_list):
            print([idx],val)
    
    elif int(menu) == 3:
        for idx, val in enumerate(score_list):
            print([idx],val)
            input_del = input("삭제할 번호를 입력해주세요 : ")
            score_list.pop(int(idx))
            print("삭제가 완료되었습니다.")
        
print("종료되었습니다.") 
```

```python
#### 성적관리 프로그램 개발

import pickle

menu = -1

with open("score_data.pkl","rb") as file:
    score_list = pickle.load(file)

while int(menu) != 0:
    menu = input("메뉴를 선택해주세요 (1 - 입력, 2 - 조회, 3 - 삭제, 0 - 종료) : ")

    if int(menu) == 1:
        name = input("이름 : ")
        math = input("수학 : ")
        science = input("과학 : ")
        eng = input("영어 : ")     

        temp = []
        
        keys = ["이름","수학","과학","영어"]
        users = [name,math,science,eng]
        dic = dict(zip(keys,users))
        temp.append(dic)    
        score_list.append(temp)
            
    elif int(menu) == 2:
        for idx, val in enumerate(score_list):
            print([idx],val)
    
    elif int(menu) == 3: #리스트가 1행밖에 나오지 않음, 삭제를 두번하는 프로세스
        for idx, val in enumerate(score_list):
            print([idx],val)
        input_del = input("삭제할 번호를 입력해주세요 : ")
        score_list.pop(int(idx))
        print("삭제가 완료되었습니다.")
        
print("종료되었습니다.") 

with open("score_data.pkl","wb") as file:
    pickle.dump(score_list,file)
```

```python
import pickle

scores = []

while True:
    menu = input("메뉴를 선택해주세요 (1 - 입력, 2 - 조회, 3 - 삭제, 0 - 종료) : ")
    if menu == "1":
        #입력
        name = input("이름 : ")
        math = int(input("수학 : "))
        science = int(input("과학 : "))
        english = int(input("영어 : "))
        
        student = {"이름":name,"수학":math,"과학":science,"영어":english}
        scores.append(student)
        
    elif menu == "2":
        #조회
        for i,user in enumerate(scores):
            print(f'[{i}] 이름 : {user["이름"]}, 수학 : {user["수학"]}, 과학 : {user["과학"]}, 영어 : {user["영어"]}')
        
    elif menu == "3":
        #삭제
        for i,user in enumerate(scores):
            print(f'[{i}] 이름 : {user["이름"]}, 수학 : {user["수학"]}, 과학 : {user["과학"]}, 영어 : {user["영어"]}')
        
        del_index = int(input("삭제할 번호를 입력해주세요 : "))
        scores.pop(del_index)
        print("삭제가 완료되었습니다.")
        
    elif menu == "0":
        #종료
        
        #파일 저장
        with open("scores.pkl","wb") as f:
            pickle.dump(scores,f)
        print("종료되었습니다.")
        break
        
```

```python
def get_inputdata():
    name = input("이름 : ")
    math = int(input("수학 : "))
    science = int(input("과학 : "))
    english = int(input("영어 : "))
    
    return name, math, science, english
```

```python
def display_all(scores):
    for i,user in enumerate(scores):
        print(f'[{i}] 이름 : {user["이름"]}, 수학 : {user["수학"]}, 과학 : {user["과학"]}, 영어 : {user["영어"]}')
```

```python
import pickle

while True:
    menu = input("메뉴를 선택해주세요 (1 - 입력, 2 - 조회, 3 - 삭제, 0 - 종료) : ")
    if menu == "1":
        #입력
        
        name, math ,science, english = get_inputdata()
        
        student = {"이름":name,"수학":math,"과학":science,"영어":english}
        scores.append(student)
        
    elif menu == "2":
        #조회
        display_all(scores)
        
    elif menu == "3":
        #삭제
        display_all(scores)
        
        del_index = int(input("삭제할 번호를 입력해주세요 : "))
        scores.pop(del_index)
        print("삭제가 완료되었습니다.")
        
    elif menu == "0":
        #종료
        
        #파일 저장
        with open("scores.pkl","wb") as f:
            pickle.dump(scores,f)
        print("종료되었습니다.")
        break
        
```
